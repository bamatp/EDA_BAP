# **Las principales causas de mortalidad de los jóvenes en España en el punto de mira**

El presente Análisis Exploratorio de Datos tiene por objeto analizar las principales causas de mortalidad de los jóvenes en España y conocer si se han ido adoptando medidas a lo largo de los últimos años para contribuir a la disminución del número de fallecimientos provocados por las mismas. 

- Las fuentes de obtención de datos han sido: INE, EPData, Statista y el Portal de Datos Abiertos el Ayuntamiento de Madrid.

- Las herramientas utilizadas han sido: VSCode, Python, Pandas, Plotly y PowerBI.

##### ***A) Análisis de las principales causas de mortalidad en los últimos 10 años de los jóvenes en España***
Para comenzar el presente análisis, era primordial conocer en primera instancia cuáles habían sido las principales causas de mortalidad en los últimos diez años. 

A priori, para conocer cuáles eran, se tuvo en consideración la muestra de todas las defunciones anuales ocasionadas en España sin descartar rango de edad alguno. 

Al encontrarnos con una población generalmente envejecida, un elevado porcentaje de defunciones correspondía a personas mayores y por ello, las principales causas de mortalidad resultantes se encontraban relacionadas con la vejez.

Por tanto, se dio un reenfoque al análisis, centrándolo únicamente en defunciones de  personas comprendidas entre los 20 y 34 años de edad.

Para llegar hasta el resultado de las dos principales causas de mortalidad en los últimos 10 años, se siguieron los siguientes pasos:

1) En primer lugar, se recopilaron diferentes datasets publicados en el [Instituto Nacional de Estadística sobre las causas y el número de fallecimientos ocasionados por año en España.](https://www.ine.es/dyngs/INEbase/es/operacion.htm?c=Estadistica_C&cid=1254736176780&menu=ultiDatos&idp=1254735573175)

2) En segundo lugar, se abrieron los archivos en VSCode con Pandas y se convirtieron en DataFrames para la posterior limpieza, transformación y análisis de los datos. 

    Para más detalle, se recomienda acudir al siguiente [archivo](https://github.com/bamatp/EDA_BAP/blob/main/src_notebooks/1_An%C3%A1lisis_principales_causas_mortalidad.ipynb), donde se encuentran explicados en el Notebook los pasos seguidos y las pruebas realizadas hasta alcanzar los resultados finales.
        
3) Por último, se llegó a la conclusión de que en los últimos 10 años, las dos principales causas de mortalidad en jóvenes han sido con una diferencia porcentual considerable y prolongada en el tiempo: el suicidio o lesiones autoinflingidas y los accidentes de tráfico.

    - Así, el suicidio ha representado entre el 14,1% y 16,3% sobre el total de las defunciones de personas comprendidas entre los 20 y 34 años, alcanzando su máximo en 2021. 
    
        Respecto al volumen total de fallecidos por esta causa, se ha situado entre 370 y 494 defunciones por año, alcanzando su máximo en 2012.

    - Adicionalmente, los accidentes de tráfico han representado entre el 10,8% y 16,5% sobre el total de las defunciones de personas comprendidas entre los 20 y 34 años, alcanzando su máximo en 2017. 
        
        Respecto al volumen total de fallecidos por esta causa, se ha situado entre 288 y 421 defunciones por año, alcanzando su máximo en 2017.

##### ***D) Conclusiones***

Tras indagar sobre las dos principales causas de mortalidad de los jóvenes en España, encontramos un punto en común entre ambas: la adicción al teléfono móvil.

- Por un lado, en la esfera de la salud mental, la gran mayoría de jóvenes afirma que no son capaces de desconectar de las redes sociales a pesar de conocer que tal "enganche" es perjudicial para su salud mental. 

    Es innegable reconocer que las redes sociales llevan consigo una serie de ventajas tales como el entretenimiento, la comunicación, la información, la publicidad, entre otras.
    
    No obstante, las redes sociales no dejan de ser un foco donde se concentran comparaciones constantes de estilo de vida que implícitamente generan un sentimiento de inferioridad entre los usuarios, ya que por ende, en los diferentes perfiles tiende a publicarse únicamente "el lado bueno de las cosas".

    Resulta alarmante que los problemas de salud mental aparezcan cada vez en edades más tempranas, en niños que a penas superan los 10 años de edad. Desde mi punto de vista, el uso masivo de dispositivos móviles y el fácil alcance de los mismos por parte de los niños, acentúan las conductas adictivas de estos dispositivos.

    Por último, a este problema se le suma que, a pesar del incremento de especialistas y predisposición de los jóvenes a recibir ayuda psicológica, desafortunadamente no todo el mundo tiene acceso a estos recursos ya que en la sanidad pública no hay suficientes medios, las listas de espera son extremadamente largas y las consultas privadas son costosas.

- Por otro lado, en la esfera de la seguridad vial, todavía no se le ha dado la importancia que se merece a las distracciones al volante por uso de dispositivos móviles. 

    Tras haber realizado el estudio de las multas impuestas en el mes diciembre en los últimos cinco años en la ciudad de Madrid, se puede observar el elevado volumen de sanciones relacionadas con el exceso de velocidad frente al de multas impuestas por el uso de dispositivos móviles durante la conducción.

    Desde mi punto de vista, la dificultad que concierna detectar un exceso de velocidad frente al uso de un dispositivo móvil es notable. Por tanto, a pesar, de que es frecuente la comisión de ambas conductas objeto de ser sancionables, el número de multas es muy dispar.

    Es cierto que últimamente la DGT ha venido endureciendo las sanciones por distracciones al volante debido a que la gran mayoría de accidentes de tráfico se producen a causa de estas, pero no ha resultado ser suficiente.

    Espero que en un futuro próximo, con el desarollo de las nuevas tecnologías consiga implantarse un sistema eficiente para la detección de estas conductas y los conductores sean conscientes de los peligros que puede ocasionar ver de forma rápida un mensaje de Whastapp o una publicación de Instragam.
